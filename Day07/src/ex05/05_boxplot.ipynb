## Import's

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

## Считываем CSV

```python
path = '../data/ab-test.csv'
df = pd.read_csv(path)
```

```python
df
```

## Построение графика

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)

boxprops = dict(color='green', linewidth=3)
medianprops = dict(color='red', linewidth=2)
whiskerprops = dict(color='red')
capprops = dict(color='red')
flierprops = dict(marker='o', color='black', alpha=0.5)

fig.suptitle("Lag before labs deadlines. After and before the Newsfeed", fontsize=15)

for i, period in enumerate(['after', 'before']):
    ax = axes[i]
    data = [
        df[(df['group'] == 'control') & (df['time'] == period)]['diff'],
        df[(df['group'] == 'test') & (df['time'] == period)]['diff']
    ]
    ax.boxplot(
        data,
        tick_labels=['control', 'test'],
        boxprops=boxprops,
        medianprops=medianprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        flierprops=flierprops,
        patch_artist=False,
        showmeans=True,
        meanprops=dict(marker='^', markerfacecolor='green', markeredgecolor='green')
    )
    ax.set_title(period)
    ax.grid(True)

plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()
```

## What was the IQR of the control group before the newsfeed?

Looking at the boxplot, we can visually estimate the interquartile range (IQR) of the *control group before the newsfeed*.

**Approximate IQR ≈ 150**

```python
control_before = df[(df['group'] == 'control') & (df['time'] == 'before')]['diff']
q1 = control_before.quantile(0.25)
q3 = control_before.quantile(0.75)
iqr = q3 - q1
print(f"IQR: {iqr:.2f}")
```
