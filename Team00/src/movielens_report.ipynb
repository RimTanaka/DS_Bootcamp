# Аналитический отчет по MovieLens
В этом Jupyter Notebook мы поочередно анализируем данные с помощью классов из модуля `movielens_analysis.py`: `Links`, `Movies`, `Tags`, `Ratings`. Каждая ячейка содержит магическую команду `%timeit`.

```python
from movielens_analysis import Links, Movies, Tags, Ratings

links = Links('ml-latest-small/links_copy.csv')
movies = Movies('ml-latest-small/movies.csv')
tags = Tags('ml-latest-small/tags.csv')
ratings = Ratings('ml-latest-small/ratings.csv')
ratings_movies = ratings.Movies(ratings.data)
ratings_users = ratings.Users(ratings.data)
```
