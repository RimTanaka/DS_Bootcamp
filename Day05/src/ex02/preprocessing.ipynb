```python
import pandas as pd
```

## Читаем файл и задаем ID индексом

```python
CSV_PATH = "/content/auto.csv"
```

```python
df = pd.read_csv(CSV_PATH, index_col="ID")
df.head()
```

## Подсчет наблюдений

```python
df.count()
```

## Удаление дубликатов

```python
df = df.drop_duplicates(subset=["CarNumber", "Make_n_model", "Fines"], keep="last")
df.count()
```

## Проверка NaN

```python
df.isna().sum()
```

## Удаляем столбцы с более чем 500 пропусков

```python
df = df.dropna(axis=1, thresh=len(df)-500)
df.isna().sum()
```

## Заполняем пропуски в Refund предыдущим значением

```python
df["Refund"] = df["Refund"].fillna(method="ffill")
df.isna().sum()
```

## Заполняем пропуски в Fines средним значением

```python
df["Fines"] = df["Fines"].fillna(df["Fines"].mean())
df.isna().sum()
```

## Разделение Make и Model, удаление Make_n_model

```python
df["Make"] = df["Make_n_model"].apply(lambda x: str(x).split()[0] if pd.notnull(x) else None)
df["Model"] = df["Make_n_model"].apply(lambda x: " ".join(str(x).split()[1:]) if pd.notnull(x) else None)

df = df.drop(columns=["Make_n_model"])
df.head()
```

## Save Json

```python
df.to_json("auto.json", orient="records", force_ascii=False, indent=4)
```
