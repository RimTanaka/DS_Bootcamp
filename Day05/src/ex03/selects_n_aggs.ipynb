```python
import pandas as pd
```

## Загружаем JSON, делаем CarNumber индексом

```python
df = pd.read_json("auto.json")
df = df.set_index("CarNumber")
df.head()
```

## Селекты

```python
df[df["Fines"] > 2100]
```

```python
df[(df["Fines"] > 2100) & (df["Refund"] == 2)]
```

```python
df[df["Model"].isin(["Focus", "Corolla"])]
```

```python
df[df.index.isin(["Y7689C197RUS", "92928M178RUS", "7788KT197RUS", "H115YO163RUS", "X758HY197RUS"])]
```

## Агрегации по Make и Model

```python
df.groupby("Make")["Fines"].median()
```

```python
df.groupby(["Make", "Model"])["Fines"].median()
```

```python
df.groupby(["Make", "Model"])["Fines"].count()
```

```python
df.groupby(["Make", "Model"])["Fines"].agg(["min", "max"])
```

```python
df.groupby(["Make", "Model"])["Fines"].std()
```

## Агрегации по CarNumber

```python
fines_count = df.groupby(df.index)["Fines"].count().sort_values(ascending=False)
fines_count
```

```python
top_car = fines_count.index[0]
df.loc[top_car]
```

```python
fines_sum = df.groupby(df.index)["Fines"].sum().sort_values(ascending=False)
fines_sum
```

```python
top_car_sum = fines_sum.index[0]
df.loc[top_car_sum]
```

```python
df.groupby(df.index)["Model"].nunique().sort_values(ascending=False)
```
