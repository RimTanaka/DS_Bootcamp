```python
import pandas as pd
```

## Загрузка данных

```python
path = "/content/feed-views.log"

df = pd.read_csv(
                path,
                sep="\t",
                skiprows=[2, 3],
                skipfooter=2,
                names=["datetime", "user"],
                index_col="datetime",
                engine='python'
                )
```
