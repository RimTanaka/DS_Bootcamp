## Импорты

```python
import pandas as pd
import sqlite3
```

## Подключаемся к БД и читаем схему таблицы

```python
conn = sqlite3.connect("/content/checking-logs.sqlite")

schema = pd.read_sql("PRAGMA table_info(pageviews);", conn)
schema
```

## Читаем 10 стр БД

```python
preview = pd.read_sql("SELECT * FROM pageviews LIMIT 10;", conn)
preview
```

## Достаем ПодТаблицу

```python
query = '''
SELECT uid, datetime
FROM pageviews
WHERE uid LIKE 'user_%'
ORDER BY uid ASC
'''
pageviews = pd.read_sql(query, conn, index_col="datetime")
pageviews.index = pd.to_datetime(pageviews.index)
pageviews.head()
```

## Закрываем соединение с БД

```python
conn.close()
```
