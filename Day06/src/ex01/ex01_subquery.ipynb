## Подключаемся к БД и читаем схему таблицы

## Подсчет, сколько строк удовлетворяют следующим условиям: status = 'ready', numTrials = 1, labnames in (laba04, laba04s, laba05, laba06, laba06s, project1)

```python
query = '''
SELECT COUNT(*) AS cnt
FROM pageviews
WHERE uid IN (
    SELECT uid
    FROM checker
    WHERE status = 'ready'
      AND numTrials = 1
      AND labname IN ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1')
)
'''
pageviews = pd.read_sql(query, conn)
pageviews.head()
```
